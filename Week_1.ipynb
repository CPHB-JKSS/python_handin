## 03 Solution part 1

```python
# 1.A
name_list = ["Christoffer", "Alex", "Benjamin", "Nicolas", "Andreas", "Henrik", "Lars", "Mie", "Helle", "Hans", "Bo", "Viktor"]
name_withH = [name for name in name_list if name.startswith("H")]

print(name_withH)
```

```python
# 1.B
toThePowerOf3 = [value **3 for value in range(1,100)]

print(toThePowerOf3) 
```

```python
# 1.C
name_list = ["Christoffer", "Alex", "Benjamin", "Nicolas", "Andreas", "Henrik", "Lars", "Mie", "Helle", "Hans", "Bo", "Viktor"]
nameTuples = [tuple((len(name),name)) for name in name_list]

print(nameTuples)
```

```python
# 1.D
str = "M4st3rm1nd"
newStr = ""
for x in range(len(str)):
    if str[x].isnumeric():
        newStr = newStr+str[x]

str2 = "".join((x for x in str if x.isnumeric()))

print(newStr)
print(str2)
```

```python
# 1.E
die1 = (1,2,3,4,5,6)
die2 = (1,2,3,4,5,6)
combos = set((x,y) for x in die1 for y in die2)

print(combos)
print("")
print(sorted(combos))
```

## 04 Solution part 2

```python
# 2.A
name_list = ["Christoffer", "Alex", "Benjamin", "Nicolas", "Andreas", "Henrik", "Lars", "Mie", "Helle", "Hans", "Bo", "Viktor"]
dict_list = {x:len(x) for x in name_list}

print(dict_list)
```

```python
# 2.B
import math
numbers = [1,2,3,20,50,250]
bigNumbers = {x:(math.sqrt(x)) for x in numbers}


print(bigNumbers)
print("")

littleNumbers = {x:round((math.sqrt(x)),2) for x in numbers}
print(littleNumbers)
```

```python
# 2.C
die1 = (1,2,3,4,5,6)
die2 = (1,2,3,4,5,6)

combos = sorted(set((x,y) for x in die1 for y in die2))
dicts = {x:round(100/len(combos),2) for x in combos}

print(dicts)
```
